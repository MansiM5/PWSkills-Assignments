App routing means mapping URL to a specific function that will handle logic for the URL. We use app routes as they bind url to the function.

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
```

```python
from flask import Flask
app= Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route("/")
def welcome_msg():
    return "Company Name: ABC Corporation \n Location: India \n Contact Detail: 999-999-9999"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

url_for() is used in Flask for URL Building. 
The first argument is the name of specified function, and then we can pass any number of keyword argument corresponding to variable part of URL.

```python
from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def welcome():
    return "Welcome to my page"

@app.route('/success/<int:score>')
def success(score):
    return "The person has passed with marks " + str(score)

@app.route('/fail/<int:score>')
def fail(score):
    return "The person has failed with marks " + str(score)

@app.route('/result/<int:marks>')
def result(marks):
    r=""
    if marks<50:
        r="fail"
    else:
        r="success"
    return redirect(url_for(result, score = marks))



if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
