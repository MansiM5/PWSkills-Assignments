Matplotlib is a visualization, graphical plotting library for Python and it's numerical extension NumPy. It is used to create numerous & diverse plots for data visualization, graphical representation & analysis purposes. It is often used as an alternative to MATLAB.
* Five Plots: scatter, bar, hist, boxplot, pie 

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
#Scatter plot is a diagram where each value in a dataset is represented by a dot.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y, c='g')
plt.xlabel("This is my x-axis")
plt.ylabel("This is my y-axis")
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Bar plot represenst a category of data with rectangular bars with length & height proportional to values that they represent.
It is used to describe the comparisons between discrete categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
```

```python
plt.barh(company, profit)
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Box plot are used to plot summary of set of data values having properties like minimum, first quartile, median, third quartile and maximum. It is used to find outliers.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
```
