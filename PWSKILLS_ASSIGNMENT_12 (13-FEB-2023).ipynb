```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

Errors that occur during numeric calculations (invalid mathematical operation) in python are defined uner ArithmeticError class. This includes ZeroDivisionError, Underflow error & Overflow error.

```python
#ZeroDivisionError occurs when any number is divided by 0
try:
    a=5/0
    print(a)
except ZeroDivisionError as e:
    print(e)
```

```python
#Underflow(Floating-point) error occurs when result of mathematical operation is too small to be represented in python's numeric types.
try:
    x=1.2-1.0
    print(x) #this is due to internal representation of floating-point number(binary)
except FloatingPointError as e: #this will be raised when a floating-point calculation will result invalid
    print(e)
    
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

LookupError class forms the base class for exceptions that are raised when an index or a key is not found for a sequence or a dictionary respectively.

```python
#IndexError
try:
    m="mansi"
    m[10]
except IndexError as e:
    print(e)
```

```python
#KeyError
try:
    d={"a":123, "b":"mansi"}
    d["c"]
except KeyError as e:
    print(e)
```
